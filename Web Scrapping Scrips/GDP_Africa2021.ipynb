```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/w/index.php?title=List_of_African_countries_by_GDP_(nominal)&oldid=1062682412'
```

```python
try:
    page = requests.get(url)
except Exception as e:
    print('Error downloading page: ',e)

soup = BeautifulSoup(page.text, 'html.parser')

# Find all tables in the webpage
tables = soup.find_all('table')
```

```python
import pandas as pd
dfs = []  # This will contain all dataframes
for table in tables:
    dfs.append(pd.read_html(str(table))[0])

gdp = dfs[0]
gdp.columns = gdp.columns.get_level_values(0)
```

```python
gdp.head()
```

```python
df = pd.DataFrame(gdp)
```

```python
df.drop('Rank', axis=1, inplace=True)
```

```python
df.columns = ['Country', 'GDP', 'GDP per capita']
```

```python
df.head()
```

```python
df.to_csv('GDP_Africa2021.csv', index=False)
#/content/GDP_Africa2022.csv
```
