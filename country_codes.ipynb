```python
import pandas as pd
import numpy as np
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.nro.net/list-of-country-codes-in-the-afrinic-region/'
try:
    page = requests.get(url)
except Exception as e:
    print('Error downloading page: ',e)

soup = BeautifulSoup(page.text, 'html.parser')

# Find all tables in the webpage
tables = soup.find_all('table')
```

```python
import pandas as pd
dfs = []  # This will contain all dataframes
for table in tables:
    dfs.append(pd.read_html(str(table))[0])

code = dfs[0]
code.columns = code.columns.get_level_values(0)


code.head()
```

```python
df = pd.DataFrame(code)
```

```python
df.drop(columns=['A 2', 'Region'], inplace=True)
```

```python
df.columns = ['Countries', 'Code']
df.head()
```

```python
df.to_csv('codes.csv', index=False)
```
