```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#import dataset
df = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/2019-2023%20Africa%5B22%5D%201.csv')
```

```python
df.head()
```

```python
#Dropping the first column
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns=['Region', 'Country', 'Airport', 'Code', 'Period', 'Total_aircraft_movements', 'International_passengers', 'Domestic_passengers', 'Direct_transit_passengers', 'Total_passengers', 'Total_cargo']
```

```python
df.head()
```

```python
#Dropping the first two rows
df.drop([1], inplace=True)
df.drop([0], inplace=True)
```

```python
df.head()
```

```python
#Splitting the Period column into Year and Month
df.Period.str.split(pat='-', expand=True)
df['Year'] = df.Period.str.split(pat='-', expand=True)[0]
df['Month'] = df.Period.str.split(pat='-', expand=True)[1]
df.drop(['Period'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
cols = ['Total_aircraft_movements', 'International_passengers', 'Domestic_passengers', 'Direct_transit_passengers', 'Total_passengers', 'Total_cargo']
for col in cols:
  df[col] = df[col].str.replace(',', '')
```

```python
df.dropna(subset=['Total_aircraft_movements'], inplace=True)
```
