```python
import pandas as pd
```

```python
pop_2019 = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/DataSets/population_2019.csv')
pop_2020 = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/DataSets/pop_Africa2020.csv')
pop_2021 = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/DataSets/pop_Africa2021.csv')
pop_2022 = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/DataSets/pop_Africa2022.csv')
pop_2023 = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/DataSets/pop_Africa2023.csv')
```

```python
codes = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/DataSets/codes.csv')
codes.head()
```

```python
pop_2019.head()
```

```python
merge = pd.merge(codes, pop_2019, on='Code')
merge.head()
```

```python
merge.drop(['Code'], axis=1, inplace=True)
merge.columns=['Country', 'Population']
merge.head()
```

```python
merge['Country'] = merge.Country.str.title()
merge.head()
```

```python
merge['Year'] = 2019
merge.head()
```

```python
pop_2020['Year'] = 2020
pop_2020.head()
```

```python
pop_2021['Year'] = 2021
pop_2021.head()
```

```python
pop_2022['Year'] = 2022
pop_2022.head()
```

```python
pop_2023['Year'] = 2023
pop_2023.head()
```

```python
df = pd.concat([merge, pop_2020, pop_2021, pop_2022, pop_2023])
df.head()
```
