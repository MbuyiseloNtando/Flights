```python

import requests
from bs4 import BeautifulSoup

#Site URL
url ='https://en.wikipedia.org/w/index.php?title=List_of_African_countries_by_population&oldid=996548756'
```

```python
try:
    page = requests.get(url)
except Exception as e:
    print('Error downloading page: ',e)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# Find all tables in the webpage
tables = soup.find_all('table')
```

```python
import pandas as pd

dfs = []  # This will contain all dataframes
for table in tables:
    dfs.append(pd.read_html(str(table))[0])

pop = dfs[0]
pop.columns = pop.columns.get_level_values(0)
```

```python
pop.head()
```

```python
df = pd.DataFrame(pop)
```

```python
df.drop( columns = ['Ranks','Source', 'Date of last figure'], inplace=True)
```

```python
df.to_csv('pop_Africa2020.csv', index=False)
```
