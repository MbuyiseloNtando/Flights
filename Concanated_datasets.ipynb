```python
import pandas as pd
import numpy as np
```

```python
#import dataset
df = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/2019-2023%20Africa%5B22%5D%201.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns=['Region', 'Country', 'Airport', 'Code', 'Period', 'Total Aircraft Movements', 'International Passengers', 'Domestic Passengers', 'Direct Transit Passengers', 'Total Passengers', 'Total Cargo']
```

```python
df.head()
```

```python
df.drop([1], inplace=True)
df.drop([0], inplace=True)
```

```python
df.head()
```

```python
#Splitting the Period column into Year and Month
df.Period.str.split(pat='-', expand=True)
df['Year'] = df.Period.str.split(pat='-', expand=True)[0]
df['Month'] = df.Period.str.split(pat='-', expand=True)[1]
df.drop(['Period'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
cols = ['Total Aircraft Movements', 'International Passengers', 'Domestic Passengers', 'Direct Transit Passengers', 'Total Passengers', 'Total Cargo']
for col in cols:
  df[col] = df[col].str.replace(',', '')
```

```python
df.isna().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df = df.astype({'Total Aircraft Movements': int, 'International Passengers': int, 'Domestic Passengers': int, 'Direct Transit Passengers': int, 'Total Passengers': int, 'Total Cargo': int, 'Year':int, 'Month':int})
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(['Region','Code'], axis=1, inplace=True)
```

```python
df.head()
```

```python
pop = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/DataSets/africa_population.csv')
gdp = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/Flights/main/DataSets/GDP_Africa.csv')
```

```python
pop.head()
pop.drop('Unnamed: 0', axis=1, inplace=True)
pop.head()
```

```python
gdp.head()
gdp.drop('Unnamed: 0', axis=1, inplace=True)
```
